# Introduction/Business Problem

Sacramento is home to a light rail system whose main customer are work commuters. For non-commuters the light rail is unpopular. To improve ridership outside of the normal work-commute, the light rail system could promote popular destinations in proximity to the light rail and provide accessible information on popular destinations near the light rail to riders. This information could also be important for people visiting Sacramento, in particular to guests to the new hotel near the University and 65th St light rail station. Light rail could be a more popular transportation mode for Sacramento residents because of the recent increase in parking rates and the increase in the amount of metered parking around the city. But many people may not realize how easy it could be to use light rail to enjoy a night in many locations around Sacramento the city and county.

My goal is to provide the Sacramento Regional Transit information on the best venues in proximity to the light rail system that could be used in marketing materials to improve the appeal of using the light rail system for new and existing riders.
